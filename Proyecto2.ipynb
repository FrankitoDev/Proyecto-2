```python
import json
import os

RUTA_JSON = "/content/tareas.json"

def cargar_tareas():
    if os.path.exists(RUTA_JSON):
        with open(RUTA_JSON, "r") as f:
            return json.load(f)
    return []

def guardar_tareas():
    with open(RUTA_JSON, "w") as f:
        json.dump(tareas, f, indent=4)

def agregar_tarea():
    titulo = input("Ingrese el tÃ­tulo de la tarea: ")
    descripcion = input("Ingrese la descripciÃ³n de la tarea: ")
    fecha_limite = input("Ingrese la fecha lÃ­mite (YYYY-MM-DD): ")

    tarea = {
        "titulo": titulo,
        "descripcion": descripcion,
        "fecha_limite": fecha_limite,
        "completada": False
    }
    tareas.append(tarea)
    guardar_tareas()
    print("âœ… Tarea agregada con Ã©xito.")

def listar_tareas():
    if not tareas:
        print("ğŸ“­ No hay tareas registradas.")
        return

    for i, tarea in enumerate(tareas):
        estado = "âœ”ï¸" if tarea["completada"] else "âŒ"
        print(f"{i + 1}. {estado} {tarea['titulo']} - {tarea['fecha_limite']}")

def completar_tarea():
    listar_tareas()
    try:
        index = int(input("Ingrese el nÃºmero de la tarea a completar: ")) - 1
        if 0 <= index < len(tareas):
            tareas[index]["completada"] = True
            guardar_tareas()
            print("âœ… Tarea marcada como completada.")
        else:
            print("âŒ NÃºmero invÃ¡lido.")
    except ValueError:
        print("âŒ Entrada invÃ¡lida.")

def eliminar_tarea():
    listar_tareas()
    try:
        index = int(input("Ingrese el nÃºmero de la tarea a eliminar: ")) - 1
        if 0 <= index < len(tareas):
            tareas.pop(index)
            guardar_tareas()
            print("ğŸ—‘ï¸ Tarea eliminada con Ã©xito.")
        else:
            print("âŒ NÃºmero invÃ¡lido.")
    except ValueError:
        print("âŒ Entrada invÃ¡lida.")

def menu():
    while True:
        print("\nğŸ“‹ GESTOR DE TAREAS")
        print("1ï¸âƒ£ Agregar tarea")
        print("2ï¸âƒ£ Listar tareas")
        print("3ï¸âƒ£ Completar tarea")
        print("4ï¸âƒ£ Eliminar tarea")
        print("5ï¸âƒ£ Salir")

        opcion = input("Seleccione una opciÃ³n: ")
        if opcion == "1":
            agregar_tarea()
        elif opcion == "2":
            listar_tareas()
        elif opcion == "3":
            completar_tarea()
        elif opcion == "4":
            eliminar_tarea()
        elif opcion == "5":
            print("ğŸ‘‹ Saliendo del gestor de tareas.")
            break
        else:
            print("âŒ OpciÃ³n no vÃ¡lida.")

tareas = cargar_tareas()
menu()
```
