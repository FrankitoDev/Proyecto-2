```python
import json
import os

RUTA_JSON = "/content/tareas.json"

def cargar_tareas():
    if os.path.exists(RUTA_JSON):
        with open(RUTA_JSON, "r") as f:
            return json.load(f)
    return []

def guardar_tareas():
    with open(RUTA_JSON, "w") as f:
        json.dump(tareas, f, indent=4)

def agregar_tarea():
    titulo = input("Ingrese el título de la tarea: ")
    descripcion = input("Ingrese la descripción de la tarea: ")
    fecha_limite = input("Ingrese la fecha límite (YYYY-MM-DD): ")

    tarea = {
        "titulo": titulo,
        "descripcion": descripcion,
        "fecha_limite": fecha_limite,
        "completada": False
    }
    tareas.append(tarea)
    guardar_tareas()
    print("✅ Tarea agregada con éxito.")

def listar_tareas():
    if not tareas:
        print("📭 No hay tareas registradas.")
        return

    for i, tarea in enumerate(tareas):
        estado = "✔️" if tarea["completada"] else "❌"
        print(f"{i + 1}. {estado} {tarea['titulo']} - {tarea['fecha_limite']}")

def completar_tarea():
    listar_tareas()
    try:
        index = int(input("Ingrese el número de la tarea a completar: ")) - 1
        if 0 <= index < len(tareas):
            tareas[index]["completada"] = True
            guardar_tareas()
            print("✅ Tarea marcada como completada.")
        else:
            print("❌ Número inválido.")
    except ValueError:
        print("❌ Entrada inválida.")

def eliminar_tarea():
    listar_tareas()
    try:
        index = int(input("Ingrese el número de la tarea a eliminar: ")) - 1
        if 0 <= index < len(tareas):
            tareas.pop(index)
            guardar_tareas()
            print("🗑️ Tarea eliminada con éxito.")
        else:
            print("❌ Número inválido.")
    except ValueError:
        print("❌ Entrada inválida.")

def menu():
    while True:
        print("\n📋 GESTOR DE TAREAS")
        print("1️⃣ Agregar tarea")
        print("2️⃣ Listar tareas")
        print("3️⃣ Completar tarea")
        print("4️⃣ Eliminar tarea")
        print("5️⃣ Salir")

        opcion = input("Seleccione una opción: ")
        if opcion == "1":
            agregar_tarea()
        elif opcion == "2":
            listar_tareas()
        elif opcion == "3":
            completar_tarea()
        elif opcion == "4":
            eliminar_tarea()
        elif opcion == "5":
            print("👋 Saliendo del gestor de tareas.")
            break
        else:
            print("❌ Opción no válida.")

tareas = cargar_tareas()
menu()
```
